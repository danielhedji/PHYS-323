## 1b

```python
def form(x,y,z):
    return (x + (y/3) + (z/2) + (11/12))
```

```python
#possible combinations for the degenerate energy states
comb1x = [[0,1,0],[1,0,0],[0,1,0],[0,1,0]]
comb1y = [[3,0,0],[1,4,1],[0,3,0],[2,2,5]]
comb1z = [[0,0,2],[0,0,2],[3,1,1],[2,0,0]]
sol = []
for i,j,k in zip(comb1x,comb1y,comb1z):
    li = []
    for f in range(len(i)):
        #applying each case into the formula for E
        s = form(i[f],j[f],k[f])
        li.append(s)
    sol.append(li)
sol
```

```python
print(np.mean(sol[0])) #E5
print(np.mean(sol[1])) #E7
print(np.mean(sol[2])) #E8
print(np.mean(sol[3])) #E9
```

## 4a

```python
def psi200(x,z):
    co = 1 / np.sqrt(8 * np.pi)
    mid = 1 - 0.5 * np.sqrt(x ** 2 + z ** 2)
    ex = np.exp(-0.5 * np.sqrt(x ** 2 + z ** 2))
    return co * mid * ex
def psi210(x,z):
    co = 1 / np.sqrt(32 * np.pi)
    mid = z 
    ex = np.exp(-0.5 * np.sqrt(x ** 2 + z ** 2))
    return co * mid * ex
def psi_squared(x,z, alpha):
    first = (4 / 5) * psi200(x,z) ** 2
    second = (1 / 5) * psi210(x,z) ** 2
    third = (2 / 5) * psi200(x,z) * psi210(x,z) * \
        np.cos(alpha)
    return first + second + third    
```

```python
x = np.linspace(-10,10,1000)
z = np.linspace(-10,10,1000)
X, Z = np.meshgrid(x,z)
```

```python
prob = psi_squared(X,Z,0)
plt.contourf(X,Z, prob, 50,cmap = "coolwarm_r")
plt.colorbar()
plt.xlabel("x")
plt.ylabel("z")
plt.title("Probability density for alpha = 0")
```

```python
prob = psi_squared(X,Z,np.pi / 4)
plt.contourf(X,Z, prob, 50,cmap = "coolwarm_r")
plt.colorbar()
plt.xlabel("x")
plt.ylabel("z")
plt.title("Probability density for alpha = pi over 4")
```

```python
prob = psi_squared(X,Z,np.pi / 3)
plt.contourf(X,Z, prob, 50,cmap = "coolwarm_r")
plt.colorbar()
plt.xlabel("x")
plt.ylabel("z")
plt.title("Probability density for alpha = pi over 3")
```

```python
prob = psi_squared(X,Z,np.pi / 2)
plt.contourf(X,Z, prob, 50,cmap = "coolwarm_r")
plt.colorbar()
plt.xlabel("x")
plt.ylabel("z")
plt.title("Probability density for alpha = pi over 2")
```

## 4b

The relative phase is causing the plots to have an increasingly more symmetric probability distribution. When there was no phase, the probability distribution was completely assymetric, and began to become more symmetric (even in both directions of z) as the phase increased.
