```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#3b
m = 9.11e-31 #kg
w = 2*np.pi* 5e15 #1/s
x_range = np.linspace(-2e-10,2e-10,100)
V_x = 0.5 * m * w ** 2 * x_range ** 2
h_bar = 6.626e-34 / (2* np.pi) #Js
#Energy states
E0 = 0.5 * w * h_bar
E1 = (1 + 0.5) * w * h_bar
E2 = (2 + 0.5) * w * h_bar
```

```python
#wave functions found in a
def psi0(m,w,h,x):
    q = (m * w / (np.pi * h)) ** 0.25
    r = np.exp(-m * w * x ** 2 / (2 * h))
    return q * r
def psi1(m,w,h,x):
    q = psi0(m,w,h,x)
    r = (2 * m * w / h) ** 0.5
    return r * q * x
def psi2(m,w,h,x):
    q = psi0(m,w,h,x)
    r = (2 ** 0.5 * h) ** -1
    s = (2 * m * w * x ** 2 - h)
    return q * r * s
```

```python
#placing the values into lists
E = [E0,E1,E2]
psi = [psi0(m,w,h_bar,x_range),psi1(m,w,h_bar,x_range),psi2(m,w,h_bar,x_range)]
```

```python
#functions meant for plotting straight lines
#each line is at the specific energy levels
plt.axhline(y = E0, label = "E0", color = "purple")
plt.axhline(y = E1, label = "E1", color = "k")
plt.axhline(y = E2, label = "E2", color = "yellow")
#plotting potential against x
plt.plot(x_range, V_x, label = "Potential")
#plotting the waveforms using the formula given
for i in range(len(E)):
    ans = E[i] + 1e-6 * E2 * psi[i]
    plt.plot(x_range,ans,label = i)
#functions to make plot look nice
plt.title("Excited states of psi in a SHO")
plt.xlabel("Position (m)")
plt.ylabel("Potential (J)")
plt.legend()
```
