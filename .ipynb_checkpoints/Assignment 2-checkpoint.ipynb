```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
x = np.linspace(0,1,1000)
pi = np.pi
L = 1
n = np.linspace(1,15,15)
en1 = []
coeff = []
inte = []
square = np.sqrt(2/L)
count = 0
while count != len(n):
    q = square * np.sin(n[count] * pi * x / L)
    en1.append(q)
    count += 1
y0 = lambda x,n: .5 * ((1.7 * (x / L) ** 2) - 0.7 * (x / L) - (x / L) ** 3) * \
np.log(1 + x / L) * square * np.sin(n * pi * x / L)
count = 0
for count in range(len(n)):
    sol = integrate.quad(y0, 0, 1, args = (n[count],))
    inte.append(sol[0])
count = 0
for i in range(len(en1)):
    c = en1[i] * inte[i]
    coeff.append(c)
```

```python
p = coeff[0]
q = coeff[0]
for i in range(1,6):
    p += coeff[i]
for j in range(1,len(coeff)):
    q += coeff[j]
```

```python
plt.plot(x,.5 * ((1.7 * (x / L) ** 2) - 0.7 * (x / L) - (x / L) ** 3) * np.log(1 + \
        x / L), label = "Exact sol")
plt.plot(x,coeff[0], label = "n = 1")
plt.plot(x,coeff[0]+coeff[1]+coeff[2], label = "n <= 2")
plt.plot(x,p, label = "n <= 5")
plt.plot(x,q, label = "n <= 15")
plt.legend()
plt.title("Approximated wave solution")
plt.xlabel("Position (m)")
plt.ylabel("Coefficients")
```
