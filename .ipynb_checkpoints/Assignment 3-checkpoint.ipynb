```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 1e-9 #nm (full width)
t = [0.,0.1e-15,0.8e-15,1.8e-15] #times given
h_bar = 6.626e-34 / (2 * np.pi) 
m = 9.11e-31
def wave(t,x,a):
    r = np.sqrt(105 * a) * x * (a - x) ** 2 / a ** 4
    q = 1 / np.exp(1 * 2 * np.pi ** 2 * h_bar * t 
        / (2 * m * a ** 2))
    return r * q
x = np.linspace(0,L,100)
for i in t:
    plt.plot(x,wave(i,x,L)**2, label = i)
plt.legend()
plt.title("Wave form with varying time")
plt.xlabel("Position (m)")
plt.ylabel("Waveform")
```
